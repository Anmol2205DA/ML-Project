```python
# Define a function to extract degree short names
def extract_degree_short_names(degree_str):
    pattern = r'([A-Z]+(?:\s*-\s*[A-Z]+)*)'
    matches = re.findall(pattern, degree_str)
    return ','.join(matches)
```

```python
df['Degree'] = df['Degree'].astype(str)  # Convert all values in the 'Degree' column to strings
df['Degree']=df['Degree'].apply(lambda degree_str: ','.join(re.findall(r'([A-Z]+(?:\s*-\s*[A-Z]+)*)', degree_str)))
```

```python
df['Degree']=df['Degree'].str.split(',').apply(lambda x: [i for i in x if len(i)>1])
```

```python
def remove_dash_and_text(specializations):
    modified_specializations = []
    for specialization in specializations:
        modified_specialization = re.sub(r'\s*-\s*.+', '', specialization)
        modified_specializations.append(modified_specialization.strip())
    return modified_specializations

# Apply the function to the 'Specialization' column using lambda and apply
df['Degree']=df['Degree'].apply(lambda x: remove_dash_and_text(x))
```

```python
# List of unique degrees in descending order of hierarchy
unique_degrees = ['PHD','MD', 'MS', 'MC', 'MDS', 'DNB', 'DDS', 'DLO', 'DCH','MPHIL',
                  'MBBS','BSMS','BASLP', 'BDS','FCPS','DGO', 'BUMS','GAMS','BSC', 'BNYS','MOT','BOT', 'MPT','MA','PGD',
                  'BA','BPT', 'BAMS', 'BHMS', 'BVS', 'AH','DHMS']

def extract_highest_degree(degrees):
    for degree in unique_degrees:
        if len(degrees)==1:
            return degrees[0]
        if any(d in degree for d in degrees):
            return degree
    return np.nan

# Apply the function to the DataFrame
df['Highest_Degree'] = df['Degree'].apply(extract_highest_degree)
```
